```python

desk=[
      ['r', 'n', 'b', 'q', 'k', 'b', 'n', 'r'],
      ['p', 'p', 'p', 'p', 'p', 'p', 'p', 'p'],
      ['.', '.', '.', '.', '.', '.', '.', '.'],
      ['.', '.', '.', '.', '.', '.', '.', '.'],
      ['.', '.', '.', '.', '.', '.', '.', '.'],
      ['.', '.', '.', '.', '.', '.', '.', '.'],
      ['P', 'P', 'P', 'P', 'P', 'P', 'P', 'P'],
      ['R', 'N', 'B', 'Q', 'K', 'B', 'N', 'R'],
     ]
help_desk=[
      ['.', '.', '.', '.', '.', '.', '.', '.'],
      ['.', '.', '.', '.', '.', '.', '.', '.'],
      ['.', '.', '.', '.', '.', '.', '.', '.'],
      ['.', '.', '.', '.', '.', '.', '.', '.'],
      ['.', '.', '.', '.', '.', '.', '.', '.'],
      ['.', '.', '.', '.', '.', '.', '.', '.'],
      ['.', '.', '.', '.', '.', '.', '.', '.'],
      ['.', '.', '.', '.', '.', '.', '.', '.'],
     ]


def print_desk(): #создаем функцию для вывода игрового поля
    print()
    print("    A B C D E F G H")
    for i in range(8):
      print(i+1,' ',end=' ')
      for j in desk[i]:
          print (j, end=' ')
      print(' ',i+1,end=' ')
      print()
    print("    A B C D E F G H")
    print("Для остановки игры введите: stop")

def name_igrok(): #создаем функцию для определения игрока
    if igrok:
        return 'Белые'
    else:
        return 'Черные'

#Возвращает фигуру по адресу из поля
def name_figura(adr_figura):
  return desk[int(adr_figura[1])-1]["ABCDEFGH".index(adr_figura[0])]

def input_figura():
  error=True #создаем переменную для проверки на ошибку
  while error: #создаем цикл, чтобы в случае ввода неправильного значения запрашивался бы еще одно
    sel_adress_figura=input(name_igrok()+ ' - выберите фигуру: ') #ввод значения
    #проверка синтаксиса
    if sel_adress_figura=="stop":
        error=False
    else:
        if len(sel_adress_figura)==2 and sel_adress_figura[1] in '12345678' and sel_adress_figura[0] in 'ABCDEFGH':
          figura=name_figura(sel_adress_figura)
          #проверка на свою фигуру
          if figura!='.' and (igrok==figura.isupper()):
            error=False
          else:
            print('Выберите свою фигуру')
        else:
          print('Не корретный адрес')
  return sel_adress_figura

def input_position(adr_figura):
  adr_position="stop"
  error=(adr_figura!="stop") #создаем переменную для проверки на ошибку
  while error: #создаем цикл, чтобы в случае ввода неправильного значения запрашивался бы еще одно
    print_help_desk(adr_figura)
    adr_position=input(name_igrok()+ ' фигура '+adr_figura+' - выберите позицию: ') #ввод значения
    #проверка синтаксиса
    if adr_position=="stop":
        error=False
    else:
        if len(adr_position)==2 and adr_position[1] in '12345678' and adr_position[0] in 'ABCDEFGH':
          figura=name_figura(adr_position)
          #проверка на отмену (ставим на место)
          if adr_figura == adr_position:
            error=False
            #Отмена хода, выбор фигуры заново
          else:
            #Проверка на правильность хода
            fig=name_figura(adr_figura)
            yfig=int('ABCDEFGH'.index(adr_figura[0]))
            xfig=int(adr_figura[1])-1
            ypos=int('ABCDEFGH'.index(adr_position[0]))
            xpos=int(adr_position[1])-1

            if fig.lower() == 'n': #конь
                if abs(xfig-xpos)==2 and abs(yfig-ypos)==1 or abs(xfig-xpos)==1 and abs(yfig-ypos)==2:
                    if desk[xpos][ypos]=='.':
                        error=False
                    elif desk[xpos][ypos]!="." and desk[xpos][ypos].islower()!=fig.islower():
                        error=False
                    else:
                        print('Конь - ход не возможен!')
                else:
                    print('Конь так не ходит!')

            if fig.lower() == 'r': #ладья
                if xfig==xpos or yfig==ypos:
                    error=False
                    if xfig==xpos:
                        ik=0
                    else:
                        ik=int((xpos-xfig)/abs(xfig-xpos))
                    if yfig==ypos:
                        jk=0
                    else:
                        jk=int((ypos-yfig)/abs(yfig-ypos))
                    i=xfig+ik
                    j=yfig+jk
                    while ((i!=xpos or j!=ypos) and not error):
                        if desk[i][j]!='.':
                            error=True
                        i+=ik
                        j+=jk
                    if desk[xpos][ypos]!="." and desk[xpos][ypos].islower()==fig.islower():
                        error=True
                    if error:
                        print('Ладья '+adr_figura+' - ход не возможен!')
                else:
                    print('Ладья так не ходит!')

            if fig.lower() == 'b': #слон
                if abs(xfig-xpos)==abs(yfig-ypos):
                    error=False
                    ik=int((xpos-xfig)/abs(xfig-xpos))
                    jk=int((ypos-yfig)/abs(yfig-ypos))
                    i=xfig+ik
                    j=yfig+jk
                    while ((i!=xpos or j!=ypos) and not error):
                        if desk[i][j]!='.':
                            error=True
                        i+=ik
                        j+=jk
                    if desk[xpos][ypos]!="." and desk[xpos][ypos].islower()==fig.islower():
                        error=True
                    if error:
                        print('Слон '+adr_figura+' - ход не возможен!')
                else:
                    print('Слон так не ходит!')

            if fig.lower() == 'q': #королева
                if abs(xfig-xpos)==abs(yfig-ypos) or xfig==xpos or yfig==ypos:
                    error=False
                    if xfig==xpos:
                        ik=0
                    else:
                        ik=int((xpos-xfig)/abs(xfig-xpos))
                    if yfig==ypos:
                        jk=0
                    else:
                        jk=int((ypos-yfig)/abs(yfig-ypos))
                    i=xfig+ik
                    j=yfig+jk
                    while ((i!=xpos or j!=ypos) and not error):
                        if desk[i][j]!='.':
                            error=True
                        i+=ik
                        j+=jk
                    if desk[xpos][ypos]!="." and desk[xpos][ypos].islower()==fig.islower():
                        error=True
                    if error:
                        print('Ферзь '+adr_figura+' - ход не возможен!')
                else:
                    print('Ферзь так не ходит!')

            if fig.lower() == 'k': #король
                if (abs(xfig-xpos)==1 and abs(yfig-ypos)==1) or (xfig==xpos and abs(yfig-ypos)==1) or (abs(xfig-xpos)==1 and yfig==ypos):
                    if desk[xpos][ypos]=='.':
                        error=False
                    elif desk[xpos][ypos].islower()!=fig.islower():
                        error=False
                    else:
                        print('Король '+adr_figura+' так не ходит! Клетка занята')
                else:
                    print('Король так не ходит!')


            if fig == 'P': #пешка белых
                if ((xfig-xpos==1 or xfig==6 and xfig-xpos==2) and yfig==ypos and desk[xpos][ypos]=='.') or (xfig-xpos==1 and abs(yfig-ypos)==1 and desk[xpos][ypos]!='.'):
                    error=False
                else:
                    print('Пешка так не ходит!')
            if fig == 'p': #пешка черных
                if ((xpos-xfig==1 or xpos-xfig==2 and xfig==1) and yfig==ypos and desk[xpos][ypos]=='.') or (xpos-xfig==1 and abs(yfig-ypos)==1 and desk[xpos][ypos]!='.'):
                    error=False
                else:
                    print('Пешка так не ходит!')

        else:
          print('Не корретный адрес')
  return adr_position

def proverka_win():
  c=0
  for i in desk:
    for j in i:
      if j.lower()=='k':
        c+=1
  if c==2:
    return False
  else:
    return True

def print_help_desk(adr_figura):
    yfig=int('ABCDEFGH'.index(adr_figura[0]))
    xfig=int(adr_figura[1])-1
    print()
    print("    A B C D E F G H       A B C D E F G H")
    for i in range(8):
      print(i+1,' ',end=' ')
      for j in range(8):
          print (desk[i][j], end=' ')
      print(' ',i+1,end=' ')
      print(' ',end=' ')
      for j in range(8):
        if yfig==j and xfig==i:
          print (desk[i][j], end=' ')
        else:
          if shek_hod(adr_figura,i,j):
            print ('.', end=' ')
          else:
            print ('?', end=' ')
      print(' ',i+1,end=' ')
      print()
    print("    A B C D E F G H       A B C D E F G H")


def shek_hod(adr_figura,xpos,ypos):
  error=True
  #Проверка на правильность хода
  fig=name_figura(adr_figura)
  yfig=int('ABCDEFGH'.index(adr_figura[0]))
  xfig=int(adr_figura[1])-1
  #ypos=int('ABCDEFGH'.index(adr_position[0]))
  #xpos=int(adr_position[1])-1

  if fig.lower() == 'n': #конь
      if abs(xfig-xpos)==2 and abs(yfig-ypos)==1 or abs(xfig-xpos)==1 and abs(yfig-ypos)==2:
          if desk[xpos][ypos]=='.':
              error=False
          elif desk[xpos][ypos].islower()!=fig.islower():
            error=False
  if fig.lower() == 'r': #ладья
      if xfig==xpos or yfig==ypos:
          error=False
          if xfig==xpos:
              ik=0
          else:
              ik=int((xpos-xfig)/abs(xfig-xpos))
          if yfig==ypos:
              jk=0
          else:
              jk=int((ypos-yfig)/abs(yfig-ypos))
          i=xfig+ik
          j=yfig+jk
          while ((i!=xpos or j!=ypos) and not error):
              if desk[i][j]!='.':
                error=True
              i+=ik
              j+=jk
          if desk[xpos][ypos]!="." and desk[xpos][ypos].islower()==fig.islower():
              error=True

  if fig.lower() == 'b': #слон
      if abs(xfig-xpos)==abs(yfig-ypos):
          error=False
          ik=int((xpos-xfig)/abs(xfig-xpos))
          jk=int((ypos-yfig)/abs(yfig-ypos))
          i=xfig+ik
          j=yfig+jk
          while ((i!=xpos or j!=ypos) and not error):
              if desk[i][j]!='.':
                  error=True
              i+=ik
              j+=jk
          if desk[xpos][ypos]!="." and desk[xpos][ypos].islower()==fig.islower():
              error=True

  if fig.lower() == 'q': #королева
      if abs(xfig-xpos)==abs(yfig-ypos) or xfig==xpos or yfig==ypos:
          error=False
          if xfig==xpos:
              ik=0
          else:
              ik=int((xpos-xfig)/abs(xfig-xpos))
          if yfig==ypos:
              jk=0
          else:
              jk=int((ypos-yfig)/abs(yfig-ypos))
          i=xfig+ik
          j=yfig+jk
          while ((i!=xpos or j!=ypos) and not error):
              if desk[i][j]!='.':
                  error=True
              i+=ik
              j+=jk
          if desk[xpos][ypos]!="." and desk[xpos][ypos].islower()==fig.islower():
              error=True

  if fig.lower() == 'k': #король
      if (abs(xfig-xpos)==1 and abs(yfig-ypos)==1) or (xfig==xpos and abs(yfig-ypos)==1) or (abs(xfig-xpos)==1 and yfig==ypos):
          if desk[xpos][ypos]=='.':
              error=False
          elif desk[xpos][ypos].islower()!=fig.islower():
              error=False

  if fig == 'P': #пешка белых
      if ((xfig-xpos==1 or xfig==6 and xfig-xpos==2) and yfig==ypos and desk[xpos][ypos]=='.') or (xfig-xpos==1 and abs(yfig-ypos)==1 and desk[xpos][ypos]!='.'):
          error=False
  if fig == 'p': #пешка черных
      if ((xpos-xfig==1 or xpos-xfig==2 and xfig==1) and yfig==ypos and desk[xpos][ypos]=='.') or (xpos-xfig==1 and abs(yfig-ypos)==1 and desk[xpos][ypos]!='.'):
          error=False
  return error


win=False #создаем переменную для отслеживания победы
igrok=False#создаем переменную для отслеживания какой игрок ходит (true-белые false-черные)
stopigra=False#для остановки игры
while win==False and not stopigra: #создаем цикл, который вызывает функции пока не кто-нибудь не победит
    igrok=not igrok #смена игрока
    adr_figura,adr_position="",""
    while adr_figura==adr_position and not stopigra:

        print_desk()
        adr_figura=input_figura()
        adr_position=input_position(adr_figura)
        if adr_figura=="stop" or adr_position=="stop":
          stopigra=True
    if not stopigra:
        desk[int(adr_position[1])-1][int('ABCDEFGH'.index(adr_position[0]))] = name_figura(adr_figura)
        desk[int(adr_figura[1])-1][int('ABCDEFGH'.index(adr_figura[0]))] = "."
        win=proverka_win()
if stopigra:
    print('Игра остановлена')
else:
  print('Выиграли '+name_igrok()) #выводим победителя

```
